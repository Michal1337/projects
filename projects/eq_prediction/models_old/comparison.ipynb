```python
def fix_f1(df):
    f1 = df['f1_score'].values
    scores = np.array([np.fromstring(s.strip('[]'), dtype=float, sep=' ') for s in f1])
    df["f1_0"] = scores[:,0]
    df["f1_1"] = scores[:,1]
    f1_val = df['val_f1_score'].values
    scores_val = np.array([np.fromstring(s.strip('[]'), dtype=float, sep=' ') for s in f1_val])
    df["f1_0_val"] = scores_val[:,0]
    df["f1_1_val"] = scores_val[:,1]
    df.drop(columns=['f1_score', 'val_f1_score'], inplace=True)
    return df
```

```python
df1 = fix_f1(df1)
df3 = fix_f1(df3)
df4 = fix_f1(df4)
df5 = fix_f1(df5)
df7 = fix_f1(df7)
df9 = fix_f1(df9)
df10 = fix_f1(df10)
```

```python
# plot f1 score
plt.plot(df1['f1_0'], label='model_v1')
plt.plot(df3['f1_0'], label='model_v3')
plt.plot(df4['f1_0'], label='model_v4')
plt.plot(df5['f1_0'], label='model_v5')
plt.plot(df7['f1_0'], label='model_v7')
plt.plot(df9['f1_0'], label='model_v9')
plt.plot(df10['f1_0'], label='model_v10')
plt.title('model f1_0')
plt.ylabel('f1_0')
plt.xlabel('epoch')
plt.legend()
plt.show()
```

```python
plt.plot(df1['f1_1'], label='model 1')
plt.plot(df3['f1_1'], label='model 2')
plt.plot(df4['f1_1'], label='model 3')
plt.plot(df9['f1_1'], label='model 4')
plt.plot(df10['f1_1'], label='model 5')
plt.title('Train F1 score')
plt.ylabel('F1 score')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
