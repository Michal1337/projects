```python
# plt.plot(df['loss'], label='train')
# plt.plot(df['val_loss'], label='validation')
plt.plot(df["output_1_loss"], label='output_1_loss')
# plt.plot(df["output_2_loss"], label='output_2_loss')
plt.plot(df["val_output_1_loss"], label='val_output_1_loss')
# plt.plot(df["val_output_2_loss"], label='val_output_2_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
def add_f1(df):
    df["f1"] = 2 * (df["precision"] * df["recall"]) / (df["precision"] + df["recall"])
    df["val_f1"] = 2 * (df["val_precision"] * df["val_recall"]) / (df["val_precision"] + df["val_recall"])
    return df
```

```python
df1 = pd.read_csv('history_model_v1.csv')
df2 = pd.read_csv('history_model_v2.csv')
df3 = pd.read_csv('history_model_v3.csv')
df4 = pd.read_csv('history_model_v4_5.csv')
df5 = pd.read_csv('history_model_v5.csv')
```

```python
df1 = add_f1(df1)
df2 = add_f1(df2)
df3 = add_f1(df3)
df4 = add_f1(df4)
df5 = add_f1(df5)
```
