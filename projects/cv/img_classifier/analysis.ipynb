```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def add_cols(df):
    df["info"] = df["info"].apply(eval)
    for key in df["info"][0].keys():
        df[key] = df["info"].apply(lambda x: x[key])
    df.drop(columns=["info"], inplace=True)
    return df
```

```python
df = pd.read_csv('results.csv', sep=';')
df = add_cols(df)
df.head()
```

```python
# boxplot with loss for each model type
df.boxplot(column='loss', by='model', figsize=(16, 9))
plt.title('Loss on test datatset for each model type', fontsize=24)
plt.suptitle('')
plt.xlabel('Model type', fontsize=20)
plt.ylabel('Loss', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(False)
plt.show()
```

```python
# boxplot with loss for each model type
df.boxplot(column='acc', by='model', figsize=(16, 9))
plt.title('Accuracy on test datatset for each model type', fontsize=24)
plt.suptitle('')
plt.xlabel('Model type', fontsize=20)
plt.ylabel('Accuracy', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(False)
plt.show()
```

```python
df.groupby('n_params')["loss"].mean().plot(y='loss', figsize=(16, 9), marker='o', linestyle='-', color='b')
plt.title('Mean loss on validation datatset for each model type', fontsize=24)
plt.suptitle('')
plt.xlabel('Number of parameters', fontsize=20)
plt.ylabel('Loss', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xscale('log')
plt.show()
```

```python
# boxplot with loss for each model type
df.boxplot(column='acc', by=['model', 'Data Augmentation'], figsize=(16, 9))
plt.title('Accuracy on test datatset for each data augmentation', fontsize=24)
plt.suptitle('')
plt.xlabel('Data augmentation type', fontsize=20)
plt.ylabel('Accuracy', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
# boxplot with loss for each model type
df.boxplot(column='acc', by='Regularization', figsize=(16, 9))
plt.title('Accuracy on test datatset for each regularization', fontsize=24)
plt.suptitle('')
plt.xlabel('Regularization type', fontsize=20)
plt.ylabel('Accuracy', fontsize=20)
plt.xticks(fontsize=16, rotation=45)
plt.yticks(fontsize=16)
plt.grid(False)
plt.show()
```
